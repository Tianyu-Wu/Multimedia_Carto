```python
import numpy as np
import pandas as pd
import os
```

```python
folder = './google_transit'
```

```python
stop_times = pd.read_csv(os.path.join(folder,'stop_times.txt'))
trips = pd.read_csv(os.path.join(folder,'trips.txt'))
calendar = pd.read_csv(os.path.join(folder,'calendar.txt'))
```

```python
stop_times
```

```python
trips
```

```python
calendar
```

```python
stop_times_day = pd.merge(stop_times, trips, on='trip_id')
stop_times_day
```

```python
service_ids = [1,2,3]
stop_times_filtered = stop_times_day[stop_times_day.service_id.isin(service_ids)]
stop_times_filtered
```

```python
frequency = stop_times_filtered.groupby(['stop_id','direction_id','service_id']).size().reset_index(name='count')
frequency
```

```python
num_weeks = 11
frequency['average'] = frequency['count']/11
```

```python
frequency
```

```python
frequency.to_csv(os.path.join(folder,'stop_frequency.txt'),index=False)
```

```python
vancouver_stops = pd.read_csv('Bus_Stops_vancouver.csv')
```

```python
stops_files = ['Bus_Stops_south_fraser.csv','Bus_Stops_richmond.csv','Bus_Stops_north_shore.csv','Bus_Stops_ne_sector.csv','Bus_Stops_burnaby.csv']
```

```python
for f in stops_files:
    stops = pd.read_csv(f)
    vancouver_stops = pd.concat([vancouver_stops, stops], ignore_index=True)
    
vancouver_boarding = vancouver_stops[["Stop_Number","Stop_Name","Sub_Region","Municipality","Avg_Daily_Boardings__Monday_Friday","Avg_Daily_Boardings__Saturday","Avg_Daily_Boardings__Sunday_Holiday","Stop_Longitude","Stop_Latitude"]]
vancouver_boarding.columns = ['stop_id','name','region','manicipality','avg_daily_boarding_mf','avg_daily_boarding_sat','avg_daily_boarding_sun','longitude','latitude']
vancouver_boarding
```

```python
vancouver_boarding.to_csv('vancouver_boarding.csv',index="False")
```
